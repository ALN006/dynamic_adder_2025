```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
latency = []
for i in range(1, 65):
    latency.append(2* math.ceil(math.log2(i)) + 2)
print(latency)
```

```python
area = []
max_adder_size = 64
for n in range(1, max_adder_size + 1):
    foo = 0
    for i in range(1, n + 1):
        for j in range(1, math.ceil(math.log2(n) + 1)):
            if(i > 2**(j)):
                foo += 3
            elif(i > 2**(j - 1)):
                foo += 2
    area.append(foo)
print(area[15])
```
